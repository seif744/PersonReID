```python
import cv2
import glob
import os
import torch
import numpy as np
```

```python
class ReIDEngine:
    def __init__(self, model, threshold=0.6):
        self.model = model
        self.model.eval()

        # store multiple features per ID
        self.gallery = {}  # {id: [feat1, feat2, ...]}
        self.next_id = 0
        self.threshold = threshold
```

```python
def preprocess(self, img):
    img = cv2.resize(img, (128, 256))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img = torch.from_numpy(img).float()
    img = img.permute(2, 0, 1) / 255.0
    img = img.unsqueeze(0)

    return img
```

```python
def extract(self, img):
    tensor = self.preprocess(img)

    with torch.no_grad():
        feat = self.model(tensor)

    feat = feat[0].cpu().numpy()

    # normalize (VERY IMPORTANT)
    feat = feat / np.linalg.norm(feat)

    return feat
```

```python
def cosine_sim(self, a, b):
    return np.dot(a, b)
```

```python
def identify(self, img):
    feat = self.extract(img)

    best_id = None
    best_score = -1

    # compare with existing gallery
    for person_id, gallery_feat in self.gallery.items():
        score = self.cosine_sim(feat, gallery_feat)

        if score > best_score:
            best_score = score
            best_id = person_id

    # CASE 1: match found
    if best_score > self.threshold:
        return best_id, best_score

    # CASE 2: new person
    new_id = self.next_id
    self.gallery[new_id] = feat
    self.next_id += 1

    return new_id, best_score
```

```python
engine = ReIDEngine(model)

paths = glob.glob("crops/**/*.jpg", recursive=True)

for p in paths:
    img = cv2.imread(p)

    if img is None:
        continue

    person_id, score = engine.identify(img)

    print(p, "→ ID:", person_id, "score:", score)
```

```python
class ReIDEngine:
    def __init__(self, model, threshold=0.6):
        self.model = model
        self.model.eval()

        # store multiple features per ID
        self.gallery = {}  # {id: [feat1, feat2, ...]}
        self.next_id = 0
        self.threshold = threshold
```

```python
def get_prototype(self, features):
    return np.mean(features, axis=0)
```

```python
def cosine_sim(self, a, b):
    return np.dot(a, b)
```

```python
def extract(self, img):
    img = cv2.resize(img, (128, 256))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img = torch.from_numpy(img).float()
    img = img.permute(2, 0, 1) / 255.0
    img = img.unsqueeze(0)

    with torch.no_grad():
        feat = self.model(img)

    feat = feat[0].cpu().numpy()
    feat = feat / np.linalg.norm(feat)

    return feat
```

```python
def identify(self, img):
    feat = self.extract(img)

    best_id = None
    best_score = -1

    # compare with PROTOTYPE of each ID
    for person_id, feats in self.gallery.items():

        prototype = self.get_prototype(feats)
        score = self.cosine_sim(feat, prototype)

        if score > best_score:
            best_score = score
            best_id = person_id

    # CASE 1: match found
    if best_score > self.threshold:
        self.gallery[best_id].append(feat)
        return best_id, best_score

    # CASE 2: new identity
    new_id = self.next_id
    self.gallery[new_id] = [feat]
    self.next_id += 1

    return new_id, best_score
```

```python
import cv2
import torch
import numpy as np


class ReIDEngine:
    def __init__(self, model, threshold=0.6, alpha=0.2):
        self.model = model
        self.model.eval()

        self.threshold = threshold
        self.alpha = alpha

        # global identity memory (tracker → global ID mapping)
        self.gallery = {}   # {global_id: prototype_embedding}
        self.next_id = 0

    # -------------------------
    # INPUT: cropped person
    # -------------------------
    def preprocess(self, img):
        img = cv2.resize(img, (128, 256))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        img = torch.from_numpy(img).float()
        img = img.permute(2, 0, 1) / 255.0
        return img.unsqueeze(0)

    # -------------------------
    # FASTREID FEATURE EXTRACTOR
    # -------------------------
    def extract(self, img):
        tensor = self.preprocess(img)

        with torch.no_grad():
            feat = self.model(tensor)[0].cpu().numpy()

        # normalize → cosine similarity becomes dot product
        feat = feat / np.linalg.norm(feat)

        return feat

    # -------------------------
    # MAIN: assign GLOBAL ID
    # -------------------------
    def identify(self, crop):
        feat = self.extract(crop)

        best_id = None
        best_score = -1

        # compare with all known identities
        for gid, proto in self.gallery.items():
            score = np.dot(feat, proto)

            if score > best_score:
                best_score = score
                best_id = gid

        # -------------------------
        # MATCH FOUND → update identity
        # -------------------------
        if best_score > self.threshold:
            old = self.gallery[best_id]

            # EMA update (smooth identity over time)
            updated = (1 - self.alpha) * old + self.alpha * feat
            self.gallery[best_id] = updated / np.linalg.norm(updated)

            return best_id, best_score

        # -------------------------
        # NEW ID
        # -------------------------
        self.gallery[self.next_id] = feat
        self.next_id += 1

        return self.next_id - 1, best_score
```
